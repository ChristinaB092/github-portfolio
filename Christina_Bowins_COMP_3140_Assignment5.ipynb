```python
class University:

  def __init__(self, name, enrollment, urban):
    self.name = str(name)
    self.enrollment = int(enrollment)
    self.urban = bool(urban)

  def get_name(self):
    return self.name

  def get_enrollment(self):
    return self.enrollment

  def get_urban(self):
    return self.urban

  def set_name(self):
    name = str(input(" "))

  def set_enrollment(self, set_enrollment):
    set_enrollment == self.enrollment
    if set_enrollment > 0:
      self.enrollment = int(input("Enter Enrollment: "))
    else:
      self.enrollment = 0
    return self.enrollment

  def set_urban(self):
    urban = bool(input())


obj1 = University("TSU",8000,True)
obj2 = University("ASU",30000,False)
print(f"{obj1.get_name()} {obj1.get_enrollment()} {obj1.get_urban()}")
print(f"{obj2.get_name()} {obj2.get_enrollment()} {obj2.get_urban()}")

obj1.set_enrollment(8500)
obj2.set_enrollment(-3500)

print(f"{obj1.get_name()} {obj1.get_enrollment()} {obj1.get_urban()}")
print(f"{obj2.get_name()} {obj2.get_enrollment()} {obj2.get_urban()}")
```
