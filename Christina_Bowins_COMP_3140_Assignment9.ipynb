```python
import pandas as pd
df = pd.read_csv('SampleAssignment9.csv')
for i in df.index:
  if df.loc[i, 'ACT'] < 33 and df.loc[i, 'Classification'] == 'Senior':
    df.drop(i, inplace = True)
print(df)
```
