```python
class University:
  def __init__(self, name, enrollment, urban):
    self.name = str(name)
    self.enrollment = int(enrollment)
    self.urban = bool(urban)

  def get_name(self):
    return self.name

  def get_enrollment(self):
    return self.enrollment

  def get_urban(self):
    return self.urban

  def set_name(self, name):
   self.name = str(name)
   return self.name

  def set_enrollment(self, enrollment):
    if enrollment > 200000 or enrollment < 0:
      print("Negative number or number above 200,000 not allowed. Try again. Enrollment was set to previous number.")
      return self.enrollment
    else:
      print("Enrollment is now:")
      self.enrollment = enrollment
      return enrollment

  def set_urban(self, urban):
    self.urban = bool(urban)
    return self.urban

class TSU(University):
  def __init__(self, location, numberOfCampuses):
    super().__init__("TSU", 8000, True)
    self.location = str(location)
    self.numberOfCampuses = int(numberOfCampuses)


  def get_location(self):
    return self.location

  def get_numberOfCampuses(self):
    return self.numberOfCampuses

  def set_location(self, location):
    self.location = location
    return self.location

  def set_numberOfCampuses(self, numberOfCampuses):
    if numberOfCampuses < 0:
      print("Negative number not allowed. Try again. Number was set to previous number.")
      return self.numberOfCampuses
    else:
      print("Number of Campuses is now:")
      self.numberOfCampuses = numberOfCampuses
      return numberOfCampuses

class Vanderbilt(University):
  def __init__(self, numberOfColleges):
    super().__init__("Vanderbilt" , 10000, True)
    self.numberOfColleges = int(numberOfColleges)

  def get_numberOfColleges(self):
    return self.numberOfColleges

  def set_numberOfColleges(self, numberOfColleges):
    if numberOfColleges < 0:
      print("Negative number not allowed. Try again. Number was set to previous number.")
      return self.numberOfColleges
    else:
      print("Number of Colleges is now:")
      self.numberOfColleges = numberOfColleges
      return numberOfColleges

tsu = TSU("Nashville", 2)
vandy = Vanderbilt(8)
print(f"name: {tsu.get_name()} enrollment: {tsu.get_enrollment()} urban: {tsu.get_urban()} location: {tsu.get_location()} numberOfCampuses: {tsu.get_numberOfCampuses()}")
print(f"name: {vandy.get_name()} enrollment: {vandy.get_enrollment()} urban: {vandy.get_urban()} numberOfColleges: {vandy.get_numberOfColleges()}")
```

```python
class Student:
  def __init__(self, name, ID):
    self.name = name
    self.ID = ID

  def display(self):
    print(f"{self.name} with {self.ID}")

class UnderGradStudent(Student):
  def __init__(self, name, ID, cls):
    Student.__init__(self, name, ID)
    self.classification = cls

  def display(self):
    print(f"{self.name} with {self.ID} and classification {self.classification}")

y = UnderGradStudent("Chris", 123, "Junior")
y.display()
```
